```python
#TIME CONVERSION to UNIX TIME
import time
import datetime
startdate= '1/1/2018'
enddate = '31/12/2019'
print(datetime.datetime.strptime(startdate, "%d/%m/%Y").timestamp())
print(datetime.datetime.strptime(enddate, "%d/%m/%Y").timestamp())
```

```python
#pull HISTORICAL DATA
#86400 denotes daily
import pandas as pd
Daily_data =polo.returnChartData('USDT_BTC', 86400, start=1514786400.0, end=1577772000.0)
Daily_data = pd.DataFrame(Daily_data)
Daily_data['True Date']= pd.to_datetime(Daily_data['date'], unit='s')
#moving average columns
Daily_data['MA10'] = Daily_data['close'].rolling(10).mean()
Daily_data['MA20'] = Daily_data['close'].rolling(20).mean()
Daily_data['MA50'] = Daily_data['close'].rolling(50).mean()
Daily_data
```
