```python
from poloniex import Poloniex
polo = Poloniex()
import os, datetime
import matplotlib.pyplot as plt
%matplotlib inline
import plotly_express as px
import datetime

#input date time and Symbol
now = datetime.datetime.now()
startdate = '1/11/2010'
end_date= now.strftime("%d/%m/%Y")
Symbol = 'USDT_BTC'

#Set the time period to use. time categories: 7200, 14400, and 86400(daily).
timeperiod = 86400


#set the moving averages to use. MA = Simple Moving Average and EMA = Exponential Moving Average
MA1 = 10
MA2 = 20
EMA1 = 10
EMA2 = 20
#Create MA Variables
MA1_var = 'MA'+str(MA1)
MA2_var = 'MA'+str(MA2)
EMA1_var = 'EMA'+str(EMA1)
EMA2_var = 'EMA'+str(EMA2)


#pull HISTORICAL DATA
#86400 denotes daily
import pandas as pd
Daily_data =polo.returnChartData(Symbol, timeperiod, start= datetime.datetime.strptime(startdate, "%d/%m/%Y").timestamp(), end= datetime.datetime.strptime(end_date, "%d/%m/%Y").timestamp())
Daily_data = pd.DataFrame(Daily_data)
Daily_data['True Date']= pd.to_datetime(Daily_data['date'], unit='s')
Daily_data['Year']=  Daily_data['True Date'].dt.year

#moving average columns
Daily_data[MA1_var] = Daily_data['close'].rolling(MA1).mean()
Daily_data[MA2_var] = Daily_data['close'].rolling(MA2).mean()
Daily_data[EMA1_var] = Daily_data['close'].ewm(span = EMA1).mean()
Daily_data[EMA2_var] = Daily_data['close'].ewm(span = EMA2).mean()

#Rules
#Shorter moving average crosses over longer. Buy at closing price Logic Buy = Short>Long or SELL = Short<Long
#The strategy will always be long or short.
#Use closing prices as entry points
#Profit/Loss -- For buy (10>50) profit = close on day of cross until until 10<50 Reversal.
#Profit/Loss -- For Sell this is the reverse

Daily_data.head(3)
```

```python
#drop the Nan to get to the start of the moving averages
Daily_data= Daily_data.dropna()
Daily_data.head(2)
```

```python
#create the long/short column:
# We create a function to loop through each row and assess whether the current trade is either Long or Short

def buy_sell_SMA (row):
   if row[MA1_var] > row[MA2_var] :
      return 'LONG'
   if row[MA1_var] < row[MA2_var] :
      return 'SHORT'    
longshort_sma = Daily_data.apply (lambda row: buy_sell_SMA(row), axis=1)
Daily_data['Long/Short SMA'] = longshort_sma

def buy_sell_EMA (row):
   if row[EMA1_var] > row[EMA2_var] :
      return 'LONG'
   if row[EMA1_var] < row[EMA2_var] :
      return 'SHORT'    
longshort_ema = Daily_data.apply (lambda row: buy_sell_EMA(row), axis=1)
Daily_data['Long/Short EMA'] = longshort_ema
Daily_data_clean = Daily_data[['True Date','Year','close','MA10','MA20','Long/Short SMA','EMA10','EMA20','Long/Short EMA']].copy()
Daily_data_clean.head(2)
```

```python
#Find where the column values change
#this function compares current value to previous value and if they are the same = True
#If current value does not equal then it is false.
#Therefore all the "false" tags will have an exit price for our long and a entry price for our short
Daily_data_clean['SMA entry and exit'] = Daily_data_clean['Long/Short SMA'].eq(Daily_data_clean['Long/Short SMA'].shift())
Daily_data_clean['EMA entry and exit'] = Daily_data_clean['Long/Short EMA'].eq(Daily_data_clean['Long/Short EMA'].shift())

Daily_data_clean
```

```python
#Lets pull out all the entry and exit points with closing price and Short or long column
EntryandExit_SMA = Daily_data_clean.loc[Daily_data_clean['SMA entry and exit'] == False]

#NEXT Calculate the profit/Loss per trade
#Step 1 Change all the rows named SHORT to Negative
EntryandExit_SMA.loc[EntryandExit_SMA['Long/Short SMA'] == 'LONG', 'close'] = EntryandExit_SMA['close']*-1
#Calculate the profit or loss on the trades. Fist column will be nan after that the profit or 
#loss is calculated on a per trade basis

#EntryandExit['Running Total'] = EntryandExit['Profit or Loss'].cumsum()
EntryandExit_SMA['SMA Profit/Loss'] =  EntryandExit_SMA.close+EntryandExit_SMA.close.shift(1) 
#Cumulative Profit or loss
EntryandExit_SMA['SMA Cumulative P/L'] =  EntryandExit_SMA['SMA Profit/Loss'].cumsum()

EntryandExit_SMA.head(10)


```

```python
#Lets pull out all the entry and exit points with closing price and Short or long column
EntryandExit_EMA = Daily_data_clean.loc[Daily_data_clean['EMA entry and exit'] == False]

#NEXT Calculate the profit/Loss per trade
#Step 1 Change all the rows named SHORT to Negative
EntryandExit_EMA.loc[EntryandExit_EMA['Long/Short EMA'] == 'LONG', 'close'] = EntryandExit_EMA['close']*-1
#Calculate the profit or loss on the trades. Fist column will be nan after that the profit or 
#loss is calculated on a per trade basis

#EntryandExit['Running Total'] = EntryandExit['Profit or Loss'].cumsum()
EntryandExit_EMA['EMA Profit/Loss'] =  EntryandExit_EMA.close+EntryandExit_EMA.close.shift(1) 
#Cumulative Profit or loss
EntryandExit_EMA['EMA Cumulative P/L'] =  EntryandExit_EMA['EMA Profit/Loss'].cumsum()
EntryandExit_EMA.head(2)
```

```python
#Group by to get profit or loss per year
profit_by_year_SMA = EntryandExit_SMA[['Year','SMA Profit/Loss']]
profit_by_year_SMA = profit_by_year_SMA.groupby(['Year']).sum()
profit_by_year_SMA

profit_by_year_EMA = EntryandExit_EMA[['Year','EMA Profit/Loss']]
profit_by_year_EMA = profit_by_year_EMA.groupby(['Year']).sum()

profit_by_year = pd.concat([profit_by_year_SMA, profit_by_year_EMA],axis=1)

profit_by_year
```

```python
profit_bar = profit_by_year[['SMA Profit/Loss','EMA Profit/Loss']].plot(kind='bar', title ="Annual Performace", figsize=(15, 10), legend=True, fontsize = 12)
```

```python
#trade statistics
total_trades_sma= len(EntryandExit_SMA)
Total_profit_points_sma = EntryandExit_SMA['SMA Profit/Loss'].sum()
largest_drawdown_sma = EntryandExit_SMA['SMA Profit/Loss'].min()
largest_winner_sma = EntryandExit_SMA['SMA Profit/Loss'].max()

total_trades_ema= len(EntryandExit_EMA)
Total_profit_points_ema = EntryandExit_EMA['EMA Profit/Loss'].sum()
largest_drawdown_ema = EntryandExit_EMA['EMA Profit/Loss'].min()
largest_winner_ema = EntryandExit_EMA['EMA Profit/Loss'].max()
```
