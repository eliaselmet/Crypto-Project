```python
#TIME CONVERSION to UNIX TIME
import time
import datetime
startdate= '1/11/2017'
enddate = '31/12/2018'

startdate = datetime.datetime.strptime(startdate, "%d/%m/%Y").timestamp()
enddate = datetime.datetime.strptime(enddate, "%d/%m/%Y").timestamp()
startdate
```

```python
#pull HISTORICAL DATA
#86400 denotes daily
import pandas as pd
Daily_data =polo.returnChartData('USDT_BTC', 86400, start= startdate, end= enddate)
Daily_data = pd.DataFrame(Daily_data)
Daily_data['True Date']= pd.to_datetime(Daily_data['date'], unit='s')
#moving average columns
Daily_data['MA10'] = Daily_data['close'].rolling(10).mean()
Daily_data['MA20'] = Daily_data['close'].rolling(20).mean()
Daily_data['MA50'] = Daily_data['close'].rolling(50).mean()
Daily_data
```
